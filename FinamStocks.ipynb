```python
%%node

function get_ids() {
    const ua = require('https')    
    
    return new Promise((resolve, reject) => {
        // let body = Buffer.from('', 'binary')
        let body = [];
        ua.get( { host: 'www.finam.ru',
                  port: 443,
                  path: '/cache/icharts/icharts.js',
                  encoding: 'binary' },
                  (resp) => {
                      resp.setEncoding('utf8')
                      resp
                      .on('data', (chunk) => {
                          // body += chunk
                          body.push(chunk)
                      })
                      .on('end', () => resolve(Buffer.concat(body)))
                  }
        )
        .on('error', (err) => reject(err.message))
        .on('close', () => {})
    })

}

async function main() {
    const vm = require('vm')
    js = await get_ids()
    
    let ctx = {}
    let _ = vm.runInNewContext(js, ctx)
    let { aEmitentIds     : id,
          aEmitentNames   : name,
          aEmitentCodes   : code,
          aEmitentMarkets : market,
          aEmitentDecp    : decp,
          aDataFormatStrs,
          aEmitentChild,
          aEmitentUrls    : url } = ctx
    /*
    let res = code.reduce((a,v,i) => {
        a[v] = { id: id[i], name: name[i] }
        return a
    }, {})
    let res = id.map((v,i,a) => {
        return [ v, code[i], name[i], url[v], decp[v], market[i]]
    })
    */
    return id.map((v,i,a) => [ v, code[i], market[i] ])
}

main().then(res => {
    finam_ids = res
    print('ok')
})
```

```python
from yarl import URL


async def download_csv(ua, _id, _ticker, date):
    
    _url = URL("http://export.finam.ru/{t}_{d:%d%m%y}_{d:%d%m%y}.csv".format(t=_ticker.upper(), d=date))
    _url = _url.update_query({ 'market': 1,                     # moex акции
                               'em': _id,                       # номер инструмента
                               'apply': 0,
                               # таймфрейм: 1 - тики, 2 - минуты; 3 - 5мин., 4 - 10мин., 7 - час, 8 - день, 9 - неделя, 10 - месяц
                               # 'p': 3,
                               'p': 1,
                               # csv fields collection: 12 - verbose, 5 - simple
                               # 'datf': 5,
                               'datf': 12,
                               'at': 1,                         # заголовки
                               'code': _ticker, 'cn': _ticker,  # тикер инструмента
                               'dtf': 3,                        # csv date format
                               'tmf': 3,                        # csv time format
                               'MSOR': 1,                       # 0 - start, 1 - end of candle
                               'df': date.day, 'dt': date.day,
                               'mf': date.month-1, 'mt': date.month-1,
                               'yf': date.year, 'yt': date.year,
                               'from': f"{date:%d.%m.%Y}", 'to': f"{date:%d.%m.%Y}",
                               'f': f"{_ticker}_{date:%d%m%y}_{date:%d%m%y}",
                               'e': '.csv',
                               'sep': 3,                        # разделитель полей
                               'sep2': 1,                       # разделитель разрядов
                               'mstime': 'on', 'mstimever': 1, })

    res = await ua.get(_url,
                       headers={ 'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:60.0) Gecko/20100101 Firefox/60.0',
                                 'Accept': '*/*',
                                 'Accept-Encoding': 'gzip, deflate',
                                 'Connection': 'close', })

    res.raise_for_status()
    # res.content is an async_generator
    _csv = await res.text()
    assert res.closed
    return _csv
```

```python
from datetime import datetime as dt
from aiohttp import ClientSession as UA, TCPConnector
from datetime import datetime as dt, timedelta as td
from finam_csv_loader import download_5min_simple
from asyncio import sleep
import pandas as pd
import io


ticker_code, ticker_id, days_num = 'TATN', 825, 1


ua = UA(connector=TCPConnector(limit=5))
cols = [ 'time','open','high','low','close','vol' ]
df = None  # pd.DataFrame([], columns=cols)

for diff in range(days_num):
    # d = dt(2019, 4, 17).date() - td(days=diff)
    d = dt.now().date() - td(days=diff)
    file = await download_5min_simple(ua, ticker_id, ticker_code, d)
    _df = pd.read_csv(io.StringIO(file),
                      header=0, sep=';', decimal='.', encoding='cp1251',
                      usecols=[1,2,3,4,5,6],
                      names=cols,)
    
    _df['time'] = _df['time'].apply(lambda t: f"{str(d)} {t}")
    _df['time'] = pd.to_datetime(_df['time'], format='%Y-%m-%d %H:%M:%S')

    if df is None:
        df = _df.copy()
    else:
        df = df.append(_df, ignore_index=True)

    await sleep(1.)

await ua.close()

df['vol'] = df['vol'].astype(float)
df.info()
```

```python
%matplotlib inline


import matplotlib.pyplot as plt


p = df.close.mean()
g = df[['vol', 'close']].groupby('close').apply(lambda _ : _).sort_values(by=['vol'], ascending=False)

plt.figure(figsize=(10,10))
plt.xlabel('Volume')
plt.ylabel('Close')
X,Y = g['vol'] / g['vol'].max(), g['close']
plt.scatter(X, Y, s=X*512)
plt.plot([p,p], color='red')
plt.grid(True)
plt.show()
# plt.close()
```
